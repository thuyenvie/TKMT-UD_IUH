## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np
from numpy import random

arr = random.randint(1000, size = 100)
print(arr)

print("\n 10 phan tu dau tien")
print(arr[: 10])

print("\n5 phan tu cuoi cung")
print(arr[-5:])

print("\nCác phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100")
ans = (arr % 5 == 0) & (arr > 100)
print(arr[ans])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
cnt = 0 #dem cac gia tri nho hon 100
i = 0
for i in range(len(arr)):
    if arr[i] < 100:
        cnt += 1
Nho_hon_100 = (arr < 100)
print(cnt)
print(arr[Nho_hon_100])
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("Tinh tong mang")
print(arr.sum())

print("\nTinh gia tri trung binh")
print(arr.mean())

print("\nTinh gia tri trung vi")
print(np.median(arr))

print("\nTinh Q1, Q2, Q3")
print("Q1 = ", np.quantile(arr, 0.25))
print("Q2 = ", np.quantile(arr, 0.5))
print("Q3 = ", np.quantile(arr, 0.75))

print("\nTim phan vi 20")
print(np.quantile(arr, 0.2))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
print('Chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên lần 1: ', random.choice(arr, size=10, replace=False))
print('Chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên lần 2: ', random.choice(arr, size=10, replace=False))
print('Chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên lần 3: ', random.choice(arr, size=10, replace=False))
print('Chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên lần 4: ', random.choice(arr, size=10, replace=False))
print('Chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên lần 5: ', random.choice(arr, size=10, replace=False))
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
print("Phần tử xuất hiện nhiều nhất: " , np.bincount(arr).argmax())
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
x = np.linspace(0, np.pi, 100)
#test
# x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

# Compute y values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the graphs
plt.plot(x, y_sin, label='y = sin(x)', linestyle='--')
plt.plot(x, y_cos, label='y = cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Đồ thị hàm y = sin(x) và y = cos(x)')
plt.legend()
plt.show()
```

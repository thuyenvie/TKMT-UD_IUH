### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
POP = np.random.randint(120,200,size = 10000)
# a> 
mean_POP = np.mean(POP)
do_lech_chuan = np.std(POP)
print("Chiều cao trung bình của quần thể : ",mean_POP)
print("Độ lệch chuẩn của quần thể : ", do_lech_chuan)
```

```python
# b>
count_180 = 0
for i in POP:
    if i >= 180:
        count_180 += 1
print("Tỷ lệ người cao trong quần thể là: ", count_180/ POP.size)
```

```python
# c>
plt.hist(POP, bins = 10,edgecolor = 'black')
plt.xlabel("Chiều cao")
plt.ylabel("Tần suất")
plt.title("Histogram về chiều cao của quần thể")
plt.show()
# theo tôi thì quần thể này có phân phối chuẩn
```

```python
#d>
plt.figure(figsize=(10,6))
plt.hist(POP, bins=30,edgecolor = 'black')

x = np.linspace(120, 200, 10000)
pdf = norm.pdf(x, loc= mean_POP, scale = do_lech_chuan)
plt.plot(x, pdf * len(POP) * (200 - 120) / 30, color='red')
plt.title('Histogram and Normal Distribution Density of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
# plt.legend()
# plt.grid(True)
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
#a>
mean_sample_20 = np.empty(100)
for i in range(0,100): 
    sample_20 = np.random.choice(POP,size = 20,replace=True)
    mean_sample = np.mean(sample_20)
    mean_sample_20[i] = mean_sample
plt.hist(mean_sample_20,bins = 10,edgecolor = 'black')
plt.xlabel("Chiều cao")
plt.ylabel("Tần suất")
plt.title("Histogram cho 100 trung bình mẫu bài 2")
plt.show()
```

```python
#b>
mean_sample_100 = np.empty(100)
for i in range(0,100): 
    sample_100 = np.random.choice(POP,size = 20,replace=True)
    mean_sample_over_100 = np.mean(sample_100)
    mean_sample_100[i] = mean_sample_over_100
plt.hist(mean_sample_100,bins = 10,edgecolor = 'black')
plt.xlabel("Chiều cao")
plt.ylabel("Tần suất")
plt.title("Histogram cho 100 trung bình mẫu bài 3")
plt.show()
```

```python
#c>
plt.figure(figsize=(12,4))
plt.hist(mean_sample_20, bins=10, color='blue', alpha=0.5, label='Sample size = 20', edgecolor='black')
plt.hist(mean_sample_100, bins=10, color='orange', alpha=0.5, label='Sample size = 100', edgecolor='black')
plt.title('Comparison of Sample Means (100 samples each)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
ti_le_nguoi_cao = []
for i in range(1000):
    sample = np.random.choice(POP,size = 100)
    tile = np.sum(sample >= 180)/100
    ti_le_nguoi_cao.append(tile)
plt.hist(ti_le_nguoi_cao, bins = 10, color = 'red',edgecolor = 'black')
plt.xlabel("Tỉ lệ")
plt.ylabel("Tần suất")
plt.title("Histogram tỷ lệ người cao của các mẫu có kích cỡ 100")
plt.show()
```

```python
#b>
ti_le_nguoi_cao2 = []
for i in range(1000):
    sample = np.random.choice(POP,size = 500)
    tile = np.sum(sample >= 180)/500
    ti_le_nguoi_cao2.append(tile)
plt.hist(ti_le_nguoi_cao2, bins = 10, color = 'red',edgecolor = 'black')
plt.xlabel("Tỉ lệ")
plt.ylabel("Tần suất")
plt.title("Histogram tỷ lệ người cao của các mẫu có kích cỡ 500")
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
A = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]
#a>
mean_A = np.mean(A)
confidence_level = 0.95
alpha = 1 - confidence_level
z = norm.ppf(1 - alpha/2)
sigma = 0.3
margin_of_error = z * sigma / np.sqrt(len(A))
confidence_interval = (mean_A - margin_of_error, mean_A + margin_of_error)
print("Khoảng tin cậy là: ",confidence_interval)
```

```python
sigma1 = np.std(A)
margin_of_error1 = z * sigma1 / np.sqrt(len(A))
confidence_interval1 = (mean_A - margin_of_error1, mean_A + margin_of_error1)
print("Khoảng tin cậy là: ",confidence_interval1)
```

c. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Tính điểm SAT trung bình của trường phổ thông đấy. (`Đáp án:  μ = 1528.74`)

d. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **99%** cho điểm SAT trung bình của trường phổ thông đấy. (`Đáp án: (1428.22, 1602.89)`)

e. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **95%** cho điểm SAT trung bình của trường phổ thông đấy. Cho biết khoảng ước lượng bạn xây dựng có đúng không, dựa trên giá trị đã tính ở câu trên.

```python
data = pd.read_csv("SAT-lard.csv")
```

```python
std_data_SAT = np.std(data['SAT Score'])
mean_data_SAT =data['SAT Score'].mean()
alpha = 0.01
n = len(data)
z = norm.ppf(1-alpha/2)
margin_of_error = z *(std_data_SAT/np.sqrt(n))
#a>
print(f"Khoang tin cay la: ({mean_data_SAT-margin_of_error},{margin_of_error+mean_data_SAT})")
```

```python
 #b>
alpha_b = 0.05
z_b = norm.ppf(1-alpha_b / 2)
margin_b = z_b * (std_data_SAT/np.sqrt(n))
print(f"Khoang tin cay la: ({mean_data_SAT-margin_b},{margin_b+mean_data_SAT})")
```

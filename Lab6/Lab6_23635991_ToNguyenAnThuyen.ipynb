### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as stats
import random
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t
```

```python
POP = np.random.uniform(120, 200, 10000)
```

```python
#a
mean_cao = np.mean(POP)
std_cao = np.std(POP)
print('Chiều cao trung bình của quần thể:', mean_cao)
print('Độ lệch chuẩn về chiều cao của quần thể:', std_cao)

#b
tall_cao = POP[POP >= 180]
print('Tỷ lệ người cao trong quần thể(>180):', len(tall_cao) / len(POP))

#c
plt.figure(figsize=(8, 6))
plt.hist(POP, bins = 10, width = 6)
plt.title('Chiều cao của quần thể')
plt.xlabel('Tần suất')
plt.ylabel('Chiều cao')
plt.show()
```

```python
#d.Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn 
#sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_cao, std_cao)
plt.plot(x, p, 'g', linewidth=2, label="Phân phối chuẩn")
plt.legend()
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
#a
n_samples = 100
sample_size_20 = 20
means_20 = []

for _ in range(n_samples):
    sample = np.random.choice(POP, size=sample_size_20, replace=False)
    sample_mean = np.mean(sample)
    means_20.append(sample_mean)

#b
sample_size_100 = 100
means_100 = []

for _ in range(n_samples):
    sample = np.random.choice(POP, size=sample_size_100, replace=False)
    sample_mean = np.mean(sample)
    means_100.append(sample_mean)


#c
plt.figure(figsize=(10, 6))
plt.hist(means_20, bins=15, edgecolor = 'black', density=True, alpha=0.7, label=f"Trung bình mẫu (n={sample_size_20})")
plt.hist(means_100, bins=15, edgecolor = 'black' , density=True, alpha=0.7, label=f"Trung bình mẫu (n={sample_size_100})")
plt.xlabel("Trung bình mẫu")
plt.ylabel("Tần suất")
plt.title("Histogram của trung bình mẫu")
plt.legend()
plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
#a
n_simulations = 1000
sample_size_100 = 100
tyLe_100 = []

for _ in range(n_simulations):
    sample = np.random.choice(POP, size=sample_size_100, replace=False)
    tall_sample = sample[sample >= 180]
    tyLe_tall = len(tall_sample) / len(sample)
    tyLe_100.append(tyLe_tall)

#b
sample_size_500 = 500
tyLe_500 = []

for _ in range(n_simulations):
    sample = np.random.choice(POP, size=sample_size_500, replace=False)
    tall_sample = sample[sample >= 180]
    tyLe_tall = len(tall_sample) / len(sample)
    tyLe_500.append(tyLe_tall)


#Vẽ histogram
plt.figure(figsize=(10, 6))
plt.hist(tyLe_100, bins=20, edgecolor = 'black' , density=True, alpha=0.7, label=f"Tỷ lệ người cao (n={sample_size_100})")
plt.hist(tyLe_500, bins=20, edgecolor = 'black' , density=True, alpha=0.7, label=f"Tỷ lệ người cao (n={sample_size_500})")
plt.xlabel("Tỷ lệ người cao")
plt.ylabel("Tần suất")
plt.title("Histogram của tỷ lệ người cao trong mẫu")
plt.legend()
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
data = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]
alpha = 0.05
# a>
sigma = 0.3
n = len(data)
x_bar = np.mean(data)

z_alpha_2 = stats.norm.ppf(1 - alpha / 2)

lower_bound = x_bar - z_alpha_2 * sigma / np.sqrt(n)
upper_bound = x_bar + z_alpha_2 * sigma / np.sqrt(n)

print("Khoảng tin cậy (biết sigma):", (lower_bound, upper_bound))

# b>
s = np.std(data, ddof=1)

t_alpha_2_n_1 = stats.t.ppf(1 - alpha / 2, n - 1)

lower_bound = x_bar - t_alpha_2_n_1 * s / np.sqrt(n)
upper_bound = x_bar + t_alpha_2_n_1 * s / np.sqrt(n)

print("Khoảng tin cậy (không biết sigma):", (lower_bound, upper_bound))
```
